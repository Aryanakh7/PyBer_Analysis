# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_city_type = pyber_data_df.groupby(["type"]).count()["ride_id"]

total_rides_city_type
```

```python
# 2. Get the total drivers for each city type
total_drivers_city_type = city_data_df.groupby(["type"]).sum()["driver_count"]

total_drivers_city_type
```

```python
#  3. Get the total amount of fares for each city type
total_fare_city_type = pyber_data_df.groupby(["type"]).sum()["fare"]

total_fare_city_type
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_ride = total_fare_city_type/total_rides_city_type

average_fare_ride
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_drivers_city = total_fare_city_type/total_drivers_city_type

average_fare_drivers_city
```

```python
#  6. Create a PyBer summary DataFrame.
PyBer_summary_city_type_df = pd.DataFrame({
    "Total Rides" : total_rides_city_type,
    "Total Drivers" : total_drivers_city_type,
    "Total Fares" : total_fare_city_type,
    "Average Fare per Ride" : average_fare_ride,
    "Average Fare per Driver" : average_fare_drivers_city})

PyBer_summary_city_type_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
PyBer_summary_city_type_df.index.name = None  

PyBer_summary_city_type_df
```

```python
#  8. Format the columns.

PyBer_summary_city_type_df["Total Rides"] = PyBer_summary_city_type_df["Total Rides"].map("{:,}".format)
PyBer_summary_city_type_df["Total Drivers"] = PyBer_summary_city_type_df["Total Drivers"].map("{:,}".format)
PyBer_summary_city_type_df["Total Fares"] = PyBer_summary_city_type_df["Total Fares"].map("${:,.2f}".format)
PyBer_summary_city_type_df["Average Fare per Ride"] = PyBer_summary_city_type_df["Average Fare per Ride"].map("${:,.2f}".format)
PyBer_summary_city_type_df["Average Fare per Driver"] = PyBer_summary_city_type_df["Average Fare per Driver"].map("${:,.2f}".format)

PyBer_summary_city_type_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

sum_fares_date_type = pyber_data_df.groupby(["date","type"]).sum()[["fare"]]

sum_fares_date_type
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()

sum_fares_date_type = sum_fares_date_type.reset_index()

sum_fares_date_type
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

sum_fares_date_type_pivot = sum_fares_date_type.pivot (index="date", columns="type", values="fare")

sum_fares_date_type_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
date_range = sum_fares_date_type_pivot.loc["2019-01-01": "2019-04-29"]

date_range
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)

sum_fares_date_type_pivot.index = pd.to_datetime(sum_fares_date_type_pivot.index)

sum_fares_date_type_pivot
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
sum_fares_date_type_pivot.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.

sum_fares_date_type_week = sum_fares_date_type_pivot.resample("W").sum()

sum_fares_date_type_week.head(10)
```

```python
# 9. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

sum_fares_date_type_week.plot(figsize=(20, 5))
plt.title("Total Fare by City Type")
plt.ylabel("Fare ($USD)")
plt.xlabel ("Month")

# Save the figure.
plt.savefig("analysis/PyBer_fare_summary.png")
plt.legend()
```
